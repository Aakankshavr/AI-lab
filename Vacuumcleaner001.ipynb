```python
class VacuumCleaner:
    def __init__(self, location, room_a, room_b):
        # Initial state of the vacuum cleaner
        self.location = location  # 'A' or 'B'
        self.room_a = room_a      # 'clean' or 'dirty'
        self.room_b = room_b      # 'clean' or 'dirty'

    def clean(self):
        # Cleaning process
        print(f"Initial state: Room A: {self.room_a}, Room B: {self.room_b}, Vacuum in Room {self.location}")

        if self.location == 'A':
            # Clean Room A if dirty
            if self.room_a == 'dirty':
                print("Cleaning Room A...")
                self.room_a = 'clean'
            # Move to Room B
            print("Moving to Room B...")
            self.location = 'B'
            # Clean Room B if dirty
            if self.room_b == 'dirty':
                print("Cleaning Room B...")
                self.room_b = 'clean'
        elif self.location == 'B':
            # Clean Room B if dirty
            if self.room_b == 'dirty':
                print("Cleaning Room B...")
                self.room_b = 'clean'
            # Move to Room A
            print("Moving to Room A...")
            self.location = 'A'
            # Clean Room A if dirty
            if self.room_a == 'dirty':
                print("Cleaning Room A...")
                self.room_a = 'clean'

        print(f"Final state: Room A: {self.room_a}, Room B: {self.room_b}, Vacuum in Room {self.location}")


# 8 possible states
states = [
    ('A', 'clean', 'clean'),
    ('A', 'clean', 'dirty'),
    ('A', 'dirty', 'clean'),
    ('A', 'dirty', 'dirty'),
    ('B', 'clean', 'clean'),
    ('B', 'clean', 'dirty'),
    ('B', 'dirty', 'clean'),
    ('B', 'dirty', 'dirty')
]

print("Aakanksha V R,1BM22CS001")
for i, state in enumerate(states):
    print(f"\n--- State {i+1} ---")
    vacuum = VacuumCleaner(state[0], state[1], state[2])
    vacuum.clean()
```
