```python
class PuzzleState:
    def __init__(self, board, zero_position, moves):
        self.board = board
        self.zero_position = zero_position  # (row, col)
        self.moves = moves  # List of moves taken to reach this state

    def is_goal(self, goal_state):
        return self.board == goal_state

    def get_neighbors(self):
        neighbors = []
        directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, Down, Left, Right
        row, col = self.zero_position

        for dr, dc in directions:
            new_row, new_col = row + dr, col + dc
            if 0 <= new_row < 3 and 0 <= new_col < 3:
                new_board = [row[:] for row in self.board]
                new_board[row][col], new_board[new_row][new_col] = new_board[new_row][new_col], new_board[row][col]
                neighbors.append(PuzzleState(new_board, (new_row, new_col), self.moves + [(new_row, new_col)]))

        return neighbors

def iterative_deepening_search(initial_state, goal_state):
    depth = 0
    while True:
        visited = set()
        result = depth_limited_search(initial_state, depth, visited, goal_state)
        if result is not None:
            return result
        depth += 1

def depth_limited_search(state, depth, visited, goal_state):
    if state.is_goal(goal_state):
        return state.moves

    if depth == 0:
        return None

    visited.add(tuple(map(tuple, state.board)))  # Add board state to visited

    for neighbor in state.get_neighbors():
        if tuple(map(tuple, neighbor.board)) not in visited:  # Check if already visited
            result = depth_limited_search(neighbor, depth - 1, visited, goal_state)
            if result is not None:
                return result

    visited.remove(tuple(map(tuple, state.board)))
    return None

def get_board_from_input(prompt):
    board = []
    print(prompt)
    for i in range(3):
        row = input(f"Enter row {i + 1} (space-separated values): ").strip().split()
        row = list(map(int, row))
        if len(row) != 3:
            raise ValueError("Each row must contain exactly 3 values.")
        board.append(row)
    return board


try:
    initial_board = get_board_from_input("Enter the initial state of the board:")
    zero_position = next((i, row.index(0)) for i, row in enumerate(initial_board) if 0 in row)

    goal_board = get_board_from_input("Enter the goal state of the board:")

    initial_state = PuzzleState(initial_board, zero_position, [])

    solution = iterative_deepening_search(initial_state, goal_board)
    if solution is not None:
        print("Solution found:", solution)
    else:
        print("No solution exists.")
except ValueError as e:
    print("Error:", e)
except Exception as e:
    print("An unexpected error occurred:", e)
```
